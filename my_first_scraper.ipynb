```python
import requests
from bs4 import BeautifulSoup as bs
import re
from io import StringIO
import pandas as pd
```

```python
#return result of request
def request_github_trending(url):
    res = requests.get(url)
    return res

url = 'https://github.com/trending'
req = request_github_trending(url)
```

```python
#find all instances of html code of repository rows and return it
def extract(page):
    soup = bs(page.text, 'html.parser')
    all_code = soup.findAll('article', class_='Box-row')
    find_all = []
    for each in all_code:
        find_all.append(each)
    return find_all

html = extract(req)
```

```python
#taking an array of all the instances of html code of the repository row.
#It will return an array of hash following this format: [{'deverloper': NAME, 'repository_name': REPOS_NAME, 'nbr_stars': NBR_STARS}, ...]
def transform(html_repos):
    res_arr = []
    for each in html_repos:
        hash_obj = {'developer': None, "repository_name": None, "nbr_stars": None}
        hash_obj['developer'] = each.h1.text.split('/')[0].strip()
        hash_obj['repository_name'] = each.h1.text.split('/')[1].strip()
        for a in each.findAll('a'):
            if (a.get('href').find("stargazers") != -1):
                hash_obj["nbr_stars"] = a.text.strip().replace(',', '')
        res_arr.append(hash_obj)
    return res_arr

repos_arr = transform(html)
repos_arr
```

```python
#taking a repository array of hash and transform it and return it into a CSV string. Each columns will be separated by , and each lines by \n
#The columns will be Developer, Repository Name, Number of Stars
def my_format(repositories_data):
    csv_str = 'Developer,Repository Name,Number of Stars\n'
    for each in repositories_data:
        string = ','.join(list(each.values()))
        csv_str += string + '\n'
    return csv_str
#if we need csv format
csv = my_format(repos_arr)
#if we need to print as DataFrame
TESTDATA = StringIO(csv)
df = pd.read_csv(TESTDATA, sep=",")
df
```
